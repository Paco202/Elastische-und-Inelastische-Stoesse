```python
from smartiS import *
import ipywidgets as widgets

from math import pi, sin, cos
from ipywidgets import HBox, VBox, Layout

class Masse(smartiS):
    
    def __init__(self):
        
        # Ebenen definieren
        layers = ["masse1",
                  "masse2",
                 "masse3"]
        lx = 650
        ly = 750

        # smartiS-Klasse initialisieren
        super().__init__(canvases = layers,
                         width    = lx,
                         height   = ly)

        # Elemente erzeugen masse1
       
        self.l_slider = widgets.IntSlider(min = 520,
                                          max = 580,
                                          value = 580,
                                          description = r"Länge $l$")
        self.l_slider.observe(self.draw_pendulum, "value")
        
        
        
        self.m_slider = widgets.IntSlider(min = 10,
                                          max = 80,
                                          value = 50,
                                          description = r"Masse $m$")
        self.m_slider.observe(self.draw_pendulum, "value")
        self.m_slider.observe(self.draw_background, "value")
        self.m_slider.observe(self.draw_winkel, "value")
        
        self.phi_slider = widgets.IntSlider(min = -45,
                                          max = 45,
                                          value = -45,
                                          description = r"Winkel $\varphi$")
        self.phi_slider.observe(self.draw_pendulum, "value")

        self.play = widgets.Play(value=70,
                                 min= 70,
                                 max=  600,
                                 step=1,
                                 interval=50,
                                 disabled=False
                                 )
        widgets.jslink((self.l_slider, 'value'), (self.play, 'value'))

        
        # Elemente erzeugen masse1
        
        self.l2_slider = widgets.IntSlider(min = 70,
                                          max = 400,
                                          value = 400,
                                          description = r"Länge $l2$")
        self.l2_slider.observe(self.draw_background, "value")
        
        self.play2 = widgets.Play(value=250,
                                 min= 30,
                                 max=  550,
                                 step=1,
                                 interval= 50,
                                 disabled= False
                                 )
        widgets.jslink((self.l2_slider, 'value'), (self.play2, 'value'))
        
        
        # Elemente erzeugen masse1
        
        self.l3_slider = widgets.IntSlider(min = 50,
                                          max = 360,
                                          value = 360,
                                          description = r"Länge $l3$")
        self.l3_slider.observe(self.draw_winkel, "value")
        
        self.play3 = widgets.Play(value=-45,
                                 min= -45,
                                 max=  45,
                                 step=1,
                                 interval= 50,
                                 disabled= False
                                 )
        widgets.jslink((self.phi_slider, 'value'), (self.play3, 'value'))
        
        # Elemente darstellen
       # self.canvas.display
      #  self.display.l_slider
      #  self.display.m_slider
       # self.display.phi_slider
      #  self.display.play
        
        
        display(HBox([
            self.canvas,
            VBox([HBox([], layout=Layout(height="600px")),
                  self.l_slider,
                  self.l2_slider,
                  self.l3_slider,
                  self.m_slider,
                  self.phi_slider,
                  self.play,
                  self.play2,
                  self.play3
                 ])
        ]))
        
        self.draw_background()
        self.draw_pendulum()
        self.draw_winkel()
       
  
    
    def draw_pendulum(self, *args):
        
        m1 = self["masse1"]
        l = self.l_slider.value
        m = self.m_slider.value
        d = m
        phi = (self.phi_slider.value + 90) * pi / 180
        #x = l * cos(phi) + 150
        x = l
        #y = l * sin(phi) +  25
        y = l
        
        with hold_canvas(m1):
        
            m1.begin_path()
            m1.move_to(150,  60)
            #p.line_to(x, y)
            m1.clear()
            m1.stroke()
            m1.stroke_arc(x, 300, d, 0, 2*pi)
            m1.fill_style = "#ff0000"
            m1.fill_arc(x, 300, d, 0, 2*pi)
  
    def draw_background(self, *args):
        
        # Masse1 zeichnen
        m2 = self["masse2"]
       
        l2 = self.l2_slider.value + 25
        m = self.m_slider.value
        d = m
        #phi = (self.phi_slider.value + 90) * pi / 180
        #x = l * cos(phi) + 150
        x = l2
        #y = l * sin(phi) +  25
        y = m
        
        with hold_canvas(m2):
            
            m2.begin_path()
            m2.move_to( 20, 75)
            #bg.line_to(275, 25)
            m2.clear()
            m2.stroke()
            m2.stroke_arc(x, 300, d, 0, 2*pi)
            m2.fill_style = "#0000ff"
            m2.fill_arc(  x, 300, d, 0, 2*pi)  
            
    def draw_winkel(self, *args):
        
        # Masse1 zeichnen
        m3 = self["masse3"]
       
        l3 = self.l3_slider.value + 25
        m = self.m_slider.value
        d = m
        phi = (self.phi_slider.value + 90) * pi / 180
        x = l3 * cos(phi) + 150
        #x = l3
        y = l3 * sin(phi) +  25
        #y = m
        
        with hold_canvas(m3):
            
            m3.begin_path()
            m3.move_to( 20, 75)
            #bg.line_to(275, 25)
            m3.clear()
            m3.stroke()
            m3.stroke_arc(x, y, d, 0, 2*pi)
            m3.fill_style = "#ff00ff"
            m3.fill_arc(  x, y, d, 0, 2*pi)
 
       
                
            
M = Masse()
```
